```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state = 69)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
import pandas as pd
```

```python
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize = (15,15), marker = 'o', hist_kwds = {'bins':20}, s=60, alpha =0.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
```

```python
print(X_new.shape)
```

```python
prediction = knn.predict(X_new)
print(prediction)
```

```python
print(iris_dataset['target_names'][prediction])
```

```python
y_pred = knn.predict(X_test)
print(y_pred)
```

```python
print(np.mean(y_pred == y_test ))
```

```python
knn.score(X_test,y_test)
```
