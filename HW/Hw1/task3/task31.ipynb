```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
#load data from CSV
data = pd.read_csv('task31.csv')
```

```python
data.head()
```

```python


#create x ticks
```

```python
fig = plt.figure(figsize=(16, 5)) 

ax = [0,0]
ax[0] = plt.gca()

science = data.USS
hanging = data.Hang_s
yr = data.Year



line1 = ax[0].plot(yr,science, c='r', label = 'US Spending on Science',marker = 'o') 

ax[1] = ax[0].twinx()

line2 = ax[1].plot(yr,hanging, c='black', label = 'Hanging Suicides',marker='o') 
lines = ax[0].get_lines() + ax[1].get_lines()  
ax[0].legend(lines,[line.get_label() for line in lines], loc='lower center')
ax[0].set_ylabel('US Spending on Science',labelpad = 12)
ax[1].set_ylabel('Hanging Suicodes',rotation = -90,labelpad = 20)
ax[0].set_title('US Spending on science, space, and technology \n correlates with \n Suiceides by hanging, strangulation and suffocation \n Correlation: 99.79% (r = 0.99789126)', fontsize=12)

y1 = range(15,35,5)
ax[0].set_yticks(y1)

y1ticks =['$15 billion', '$20 billion', '$25 billion', '$30 billion']
ax[0].set_yticklabels(y1ticks)

y2 = range(4000,12000,2000)
ax[1].set_yticks(y2)

y2ticks =['4000 suicides', '6000 suicides', '8000 suicides', '10000 suicides']
ax[1].set_yticklabels(y2ticks)

x_ticks = range(1999,2010)
ax[0].set_xticks(x_ticks)
plt.tight_layout()

plt.show()
fig.savefig('task31',bbox = 'tight')
```
