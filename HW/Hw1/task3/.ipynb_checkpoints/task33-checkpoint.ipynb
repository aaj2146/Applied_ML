```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston
```

```python
print("Shape of boston data: {}".format(boston.data.shape),len(boston.target))
print(len(boston.data[:,0]),len(boston.target))
```

```python
boston.feature_names
```

```python
fig,ax = plt.subplots(13,1, figsize=(5,20))

for i in range(13):
    ax[i].scatter(boston.target,boston.data[:,i], alpha = 0.2)
    ax[i].set_ylabel(boston.feature_names[i])

ax[12].set_xlabel('MEDV')

plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_boston
boston = load_boston()

fig,ax = plt.subplots(13,1, figsize=(5,20))

for i in range(13):
    ax[i].scatter(boston.target,boston.data[:,i], alpha = 0.2)
    ax[i].set_ylabel(boston.feature_names[i])

ax[12].set_xlabel('MEDV')

plt.tight_layout()
plt.savefig('task33.png')
plt.show()
```
