```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
#load data from CSV
data = pd.read_csv('task31.csv')
```

```python
data.head()
```

```python
#assigning columsn to variables
science = data['US_Science_spending']
hanging = data['Hanging_suicides']
yr = data['Year']

#creating y1 ticks and y2ticks using for loop

y1ticklables =[]
y2ticklables =[]
y1 = range(15,35,5)
y2 = range(4000,12000,2000)
for i in range(4):
    a = '$' + str(y1[i]) + ' billion'
    y1ticklables.append(a)
    b = str(y2[i]) +' suicides'
    y2ticklables.append(b)
print (y1ticklables)
print (y2ticklables)

#create x ticks
x = range(1999,2010)
```

```python
#using twinx function to plot 2 lines with 2 diff y-axes 
plt.figure(figsize=(8, 6))
ax1 = plt.gca()
line1 = ax1.plot(yr,science, c='r', label = 'US Spending on Science') #create label to be shown in legend
ax2 = ax1.twinx()
line2 = ax2.plot(yr,hanging, c='b', label = 'Hanging Suicides') 
lines = ax1.get_lines() + ax2.get_lines()  #create a list of lines
ax1.legend(lines,[line.get_label() for line in lines], loc='upper left')
ax1.set_ylabel('US Spending in Science')
ax2.set_ylabel('Hanging Suicodes',rotation = -90,labelpad = 15) #labelpad gives gap to ylable from axis
ax1.set_title('US Spending on science, space, and technology \n correlates with \n Suiceides by hanging, strangulation and suffocation', fontsize=12)

#assign yticks and yticklabels - use object oriented interface set_ commands
ax1.set_yticks(y1)
ax1.set_yticklabels(y1ticklables)

ax2.set_yticks(y2)
ax2.set_yticklabels(y2ticklables)

ax1.set_xticks(x)
plt.tight_layout()
plt.show()
```
