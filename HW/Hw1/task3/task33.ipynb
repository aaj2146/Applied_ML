```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_boston
boston = load_boston()

print("Keys of boston dataset: \n{}".format(boston.keys()))

print("Feature names: \n{}".format(boston['feature_names']))

fig,ax = plt.subplots(13,1, figsize = (6,60))
labels1 = ['per capita crime rate by town','proportion of residential land \n zoned for lots over 25,000 sq.ft.','proportion of non-retail \n business acres per town','Charles River \n dummy variable','nitric oxides concentration \n (parts per 10 million)','average number of \n rooms per dwelling','proportion of owner-occupied \n units built prior to 1940','weighted distances \n to five Boston \n employment centres','index of accessibility to \n radial highways','full-value property-tax rate per $10,000','pupil-teacher ratio by town','proportion of blacks by town','lower status of the population','Median value of owner-occupied homes in $1000s']
labels2 = np.array(labels1)
for i in range(13):
    ax[i].scatter( boston.target,boston.data[:,i], alpha = 0.2)
    ax[i].set_xlabel('MEDV')
    ax[i].set_ylabel(labels2[i])

    
ax[0].set_title('Boston housing dataset scatter plot', fontsize=12)

plt.tight_layout()
plt.savefig('task33.png')
plt.show()
```
