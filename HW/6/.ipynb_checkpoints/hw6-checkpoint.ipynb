# task 1

```python
iris = datasets.load_iris()
```

```python
target_values = list(set(iris.target))
num_classes = len(target_values)
len(iris.data[0])
```

```python
X_train_val,X_test, y_train_val, y_test = train_test_split(iris.data, iris.target, test_size=0.1, random_state=42,
                                                           stratify =  iris.target, shuffle = True)
```

```python
print (X_train_val.shape)
print (X_test.shape)
print (y_train_val.shape)
print (y_test.shape)
```

```python
sss = StratifiedShuffleSplit(n_splits=1, train_size=.8, test_size=.2, random_state=42)
scaler = StandardScaler()
```

```python
X_train_val_scaled = scaler.fit_transform(X_train_val)
X_test_scaled = scaler.transform(X_test)
#convert class vectors to binary class matrices
y_train_val = keras.utils.to_categorical(y_train_val, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier, KerasRegressor
from sklearn.model_selection import GridSearchCV

def make_model(optimizer="adam", hidden_size1=32, hidden_size2=32):
    model = Sequential([
        Dense(hidden_size1, input_shape=(4,), activation='relu'),#,kernel_regularizer=regularizers.l2(reg1)),
        Dense(hidden_size2, activation='relu'),#,kernel_regularizer=regularizers.l2(reg1)),
        Dense(3, activation='softmax')
    ])
    model.compile(optimizer = optimizer, loss = "categorical_crossentropy", metrics=['accuracy'])
    return model
```

```python
clf = KerasClassifier(make_model)
param_grid = {'hidden_size1': [8,12,32,64],'hidden_size2': [8,12,32,64],'epochs': [1, 5, 10]}
             #'reg1':np.logspace(-3,3,7)}
grid = GridSearchCV(clf, param_grid=param_grid, cv=sss)
grid.fit(X_train_val_scaled, y_train_val)
```

```python
grid.score(X_test_scaled,y_test)
```

```python
grid.best_params_
```

```python
res = pd.DataFrame(grid.cv_results_)
res.pivot_table(index=["param_epochs", "param_hidden_size1"],
                values=['mean_train_score', "mean_test_score"])


```

# task 2

```python
from keras.datasets import mnist
data = mnist
num_classes = 10
(x_trainval, y_trainval), (x_test, y_test) = mnist.load_data()
```

```python
cv_ind = np.random.choice(60000, 10000, replace=False)
x_cv = x_trainval[cv_ind]
y_cv = y_trainval[cv_ind]
```

```python
train_ind = [x for x in range(0,60000) if x not in cv_ind]
```

```python
x_train = x_trainval[train_ind]
y_train = y_trainval[train_ind]
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes)
y_cv = keras.utils.to_categorical(y_cv, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
x_train = x_train.reshape(50000,784)
x_cv = x_cv.reshape(10000,784)
x_test = x_test.reshape(10000,784)
```

```python
x_train = x_train/255
x_cv = x_cv/255
x_test = x_test/255
```

```python
model = Sequential([
    Dense(512, input_shape=(784,), activation='relu'),
    Dense(512, activation = 'relu'),
    Dense(512, activation = 'relu'),
    Dense(512, activation= 'relu'),
    Dense(10, activation='softmax')])
model.compile("adam", "categorical_crossentropy", metrics=['accuracy'])
```

```python
 vanilla_model = model.fit(x_train, y_train,validation_data=(x_cv,y_cv), batch_size=128, epochs=10, verbose=1)
```

```python
score = model.evaluate(x_cv, y_cv, verbose=0)
print("CV loss: {:.3f}".format(score[0]))
print("CV Accuracy: {:.3f}".format(score[1]))
```

```python
from keras.layers import Dropout
model2 = Sequential([
    Dense(512, input_shape=(784,), activation='relu'),
    Dropout(.5),
    Dense(512, activation = 'relu'),
    Dropout(.5),
    Dense(512, activation = 'relu'),
    Dropout(.5),
    Dense(512, activation= 'relu'),
    Dropout(.5),
    Dense(512, activation = 'relu'),
    Dropout(.5),
    Dense(512, activation= 'relu'),
    Dropout(.5),
    Dense(10, activation='softmax')])
model2.compile("adam", "categorical_crossentropy", metrics=['accuracy'])
```

```python
 dropout_model = model2.fit(x_train, y_train, validation_data=(x_cv,y_cv),batch_size=128, epochs=10, verbose=1)
```

```python
score2 = model2.evaluate(x_cv, y_cv, verbose=0)
print("CV loss: {:.3f}".format(score2[0]))
print("CV Accuracy: {:.3f}".format(score2[1]))
```

```python
def plot_history(logger):
    df = pd.DataFrame(logger.history)
    df[['acc', 'val_acc']].plot()
    plt.ylabel("accuracy")
    df[['loss', 'val_loss']].plot(linestyle='--', ax=plt.twinx())
    plt.ylabel("loss")
```

```python
plot_history(vanilla_model)
```

```python
plot_history(dropout_model)
```

dropout model seems to be training in a uniform manner

```python
#final test score using droput model
score_test = model2.evaluate(x_test, y_test, verbose=0)
print("CV loss: {:.3f}".format(score_test[0]))
print("CV Accuracy: {:.3f}".format(score_test[1]))
```
