```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import fetch_california_housing
import seaborn as sns
sns.set(color_codes=True)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
cal_data = fetch_california_housing()
```

```python
#cal_data
#median house value, median income, housing median age, total rooms, 
#total bedrooms, population, households, latitude, and lo ngitude in that order.
titles = ['Median Income', 'Housing Median Age', 'Average Rooms', 'Average Bed Rooms', 'Population', 'Average Occupancy',
'Latitude','Longitude']
```

```python
cal = pd.DataFrame(cal_data.data, columns = cal_data.feature_names)
avg_house_val = pd.DataFrame(cal_data.target, columns = ['avg_house_value'])
cal.head()
```

```python
cal.columns[1]
```

```python
#plotting univariate distributions for all variables in the dataset
fig,ax = plt.subplots(3,3, figsize=(15,15)) #initiating 3*3 grid of plots 

k = 0
for i in range(3):
    for j in range(3):
        if i == 2 and j == 2:
            sns.distplot(avg_house_val.iloc[:,0],ax = ax[i,j])
            break
        sns.distplot(cal.iloc[:,k],ax = ax[i,j])
        k +=1 
```

* Average occupancy seems to have outliers

1.3 Split data in training and test set.
Evaluate Linear Regression (OLS), Ridge, Lasso and
ElasticNet using cross-validation with the default parameters.

```python
X_train, X_test, y_train, y_test = train_test_split(cal, avg_house_val, random_state=0)

lr_score = np.mean(cross_val_score(LinearRegression(), X_train, y_train))
r_score = np.mean(cross_val_score(Ridge(), X_train, y_train))
l_score = np.mean(cross_val_score(Lasso(), X_train, y_train))
en_score = np.mean(cross_val_score(ElasticNet(), X_train, y_train))

print ('Mean CV score - Linear Regression (OLS) = {:.3f}'.format(lr_score))
print ('Mean CV score - Ridge Regression = {:.3f}'.format(r_score))
print ('Mean CV score - Lasso Regression = {:.3f}'.format(l_score))
print ('Mean CV score - Elastic Net = {:.3f}'.format(en_score))
```

Does scaling the data with
StandardScaler help?

```python
scaler = StandardScaler()
scaler.fit(X_train) #fitting scaler
X_train_scaled = scaler.transform(X_train) #transforming train data

X_test_scaled = scaler.transform(X_test) #transforming test data

lr_score = np.mean(cross_val_score(LinearRegression(), X_train_scaled, y_train))
r_score = np.mean(cross_val_score(Ridge(), X_train_scaled, y_train))
l_score = np.mean(cross_val_score(Lasso(), X_train_scaled, y_train))
en_score = np.mean(cross_val_score(ElasticNet(), X_train_scaled, y_train))

print ('Mean CV score - Linear Regression (OLS) on Scaled data = {:.3f}'.format(lr_score))
print ('Mean CV score - Ridge Regression on Scaled data = {:.3f}'.format(r_score))
print ('Mean CV score - Lasso Regression on Scaled data = {:.3f}'.format(l_score))
print ('Mean CV score - Elastic Net on Scaled data = {:.3f}'.format(en_score))
```

Scaling the data worsened the scores

## 1.4
Tune the parameters of the models using GridSearchCV. Do the results improve?
Visualize the dependence of the validation score on the parameters for Ridge, Lasso and
ElasticNet.

```python
param_grid_r = {'alpha': np.logspace(-3, 3, 40)}
param_grid_l = {'alpha': np.logspace( -3, 0,40)}
param_grid_en = {'alpha': np.logspace(-4, -1, 10),'l1_ratio': [0.01, .1, .5, .9, .98, 1]}

print ('Ridge:',param_grid_r,'\nLasso:',param_grid_l,'\nElastic Net:',param_grid_en)
```

```python
grid_r = GridSearchCV(Ridge(), param_grid_r)
grid_l = GridSearchCV(Lasso(), param_grid_l)
grid_en = GridSearchCV(ElasticNet(), param_grid_en)

grid_r.fit(X_train,y_train)
grid_l.fit(X_train,y_train)
grid_en.fit(X_train,y_train)

print('Ridge Regression - Best Parameters :',grid_r.best_params_)
print('Ridge Regression - Best Score :{:.4f}'.format(grid_r.best_score_))

print('\nLasso Regression - Best Parameters :',grid_l.best_params_)
print('Lasso Regression - Best Score :{:.4f}'.format(grid_l.best_score_))

print('\nElastic Net Regression - Best Parameters :',grid_en.best_params_)
print('Elastic Net Regression - Best Score :{:.4f}'.format(grid_en.best_score_))
```

```python
#ridge results dataframe
results_r = pd.DataFrame(grid_r.cv_results_)
#results_r.columns
```

```python
plt.plot(results_r['param_alpha'],results_r['mean_train_score'],label = 'Mean train score')
plt.plot(results_r['param_alpha'],results_r['mean_test_score'],label = 'Mean test score')
plt.legend()
plt.title('Ridge Regression validation score vs Alpha')
plt.xlabel('Alpha (log-scale)')
plt.ylabel('Ridge Regression Validation Score')
plt.xscale('log')
```

```python
#ridge results dataframe
results_l = pd.DataFrame(grid_l.cv_results_)
#results_l.head()
```

```python
plt.plot(results_l['param_alpha'],results_l['mean_train_score'],label = 'Mean train score')
plt.plot(results_l['param_alpha'],results_l['mean_test_score'],label = 'Mean test score')
plt.legend()
plt.title('Lasso Regression validation score vs Alpha')
plt.xlabel('Alpha (log-scale)')
plt.ylabel('Lasso Regression Validation Score')
plt.xscale('log')
```

```python
results_en = pd.DataFrame(grid_en.cv_results_)
#results_en.head()
```

```python
res = pd.pivot_table(pd.DataFrame(grid_en.cv_results_),
                     values='mean_test_score', index='param_alpha', columns='param_l1_ratio')
#res
```

```python
a4_dims = (8, 8)
fig, ax = plt.subplots(figsize=a4_dims)
sns.heatmap(ax = ax, data = res, annot=True)
ax.set_yticklabels(np.round(param_grid_en['alpha'],4))
plt.show()
```
